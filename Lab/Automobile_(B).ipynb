```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# New Section

```python
df = pd.read_csv("autompg.csv")
```

```python
print(df.columns)
```

```python
print(df.head())
```

```python
plt.figure(figsize=(12, 6))
df[['mpg', 'cylinders', 'horsepower', 'displacement', 'model year', 'car name']].hist(bins=10, figsize=(12, 8))
plt.suptitle('Distribution of Continuous Variables', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['horsepower'], y=df['acceleration'])
plt.title('Relationship between horsepower and acceleration')
plt.xlabel('horsepower')
plt.ylabel('acceleration')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x=df['model year'])
plt.title('Frequency of model year')
plt.xlabel('model year')
plt.ylabel('origin')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.pointplot(x=df['origin'], y=df['mpg'])
plt.title('City origin by mpg')
plt.xlabel('origin')
plt.ylabel('mpg')
plt.show()
```
